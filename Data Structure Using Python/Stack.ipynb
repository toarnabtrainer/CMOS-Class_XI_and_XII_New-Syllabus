```python
# stack implementation
class Stack:
    def __init__(self, max_size):
        self.__max_size = max_size
        self.__elements = [None] * self.__max_size
        self.__top = -1
    def get_max_size(self):
        return self.__max_size
    def is_full(self):
        if (self.__top == (self.__max_size - 1)):
            return True
        return False
    def is_empty(self):
        if (self.__top == -1):
            return True
        return False
    def push(self, data):
        if (self.is_full()):
            print ("O V E R F L O W ...")
        else:
            self.__top += 1
            self.__elements[self.__top] = data
    def pop(self):
        if (self.is_empty()):
            print ("U N D E R F L O W ...")
        else:
            data = self.__elements[self.__top]
            self.__top -= 1
            return data
    def display(self):
        if (self.is_empty()):
            print ("The stack is empty...")
        else:
            index = self.__top
            while (index >= 0):
                print (self.__elements[index])
                index -= 1
    # overriding __str__() in-built Python method which returns
    # one str class object
    def __str__(self):   
        msg = []
        index = self.__top
        while (index >= 0):
            msg.append((str)(self.__elements[index]))
            index -= 1
        msg = " ".join(msg)
        msg = "Stack content (Top to Bottom): " + msg
        return msg
    def find_average(self):
        # calculate the average of all elements in the stack and
        # push it to the stack
        total = 0
        position = self.__top
        if (self.is_empty()):
            print ("The Stack is empty...")
        else:
            while (position >= 0):
                total = total + self.__elements[position]
                position -= 1
            self.push(total / (self.__top + 1))
# Now creating Stack class object and calling different methods...
num_stack = Stack(8)
num_list = [11, 22, 33, 44, 55, 66]
for item in num_list:
    num_stack.push(item)
# num_stack.display()
print (num_stack)
num_stack.find_average()
print (num_stack)
```

```python
# infix to postfix conversion
class Stack:
    def __init__(self, max_size):
        self.__max_size = max_size
        self.__elements = [None] * self.__max_size
        self.__top = -1
    def get_max_size(self):
        return self.__max_size
    def is_full(self):
        if (self.__top == (self.__max_size - 1)):
            return True
        return False
    def is_empty(self):
        if (self.__top == -1):
            return True
        return False
    def peek(self):
        if(self.is_empty()):
            return None
        else:
            return self.__elements[self.__top]
    def push(self, data):
        if (self.is_full()):
            print ("O V E R F L O W ...")
        else:
            self.__top += 1
            self.__elements[self.__top] = data
    def pop(self):
        if (self.is_empty()):
            print ("U N D E R F L O W ...")
        else:
            data = self.__elements[self.__top]
            self.__top -= 1
            return data
    def __str__(self):   
        msg = []
        index = self.__top
        while (index >= 0):
            msg.append((str)(self.__elements[index]))
            index -= 1
        msg = " ".join(msg)
        msg = "Stack content (Top to Bottom): " + msg
        return msg
def infix_to_postfix(infixstr):
    infixstr += ')'
    stk = Stack(len(infixstr))
    stk.push('(')
    precedence = {"(":0, "+":1, "-":1, "*":2, "/":2}
    postfixstr = ""
    for symbol in infixstr:
        if(symbol == "("):
            stk.push(symbol)
        elif(symbol.isdigit() or symbol.isalpha()):
            postfixstr += symbol
        elif(symbol == ")"):
            while(stk.peek() != "("):
                postfixstr += stk.pop()
            stk.pop()
        else:
            while (precedence[stk.peek()] > precedence[symbol]):
                postfixstr += stk.pop()
            stk.push(symbol)
    return postfixstr
print (infix_to_postfix("a+(b/c)*d-(e*f)"))
```
