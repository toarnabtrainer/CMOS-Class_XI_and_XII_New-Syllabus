```python
# circular queue implementation
class Queue:
    def __init__(self, max_size):
        self.__max_size = max_size
        self.__elements = [None] * self.__max_size
        self.__front = -1
        self.__rear = -1
    def get_max_size(self):
        return self.__max_size
    def is_full(self):
        # write code here. If full return True other False
        if ((self.__front == 0 and self.__rear == self.__max_size-1) or
           (self.__front == self.__rear + 1)):
            return True
        else:
            return False
    def is_empty(self):
        # write code here. If empty return True other False
        if (self.__front == -1):
            return True
        else:
            return False
    def enqueue(self, data):
        if (self.is_full()):
            print ("O V E R F L O W ...")
        else:
            if (self.__front == -1):
                self.__front = 0
                self.__rear = 0
            else:
                self.__rear = (self.__rear + 1) % self.__max_size
            self.__elements[self.__rear] = data
    def dequeue(self):
        if (self.is_empty()):
            print ("U N D E R F L O W ...")
        else:
            data = self.__elements[self.__front]
            if (self.__front == self.__rear):
                self.__front = -1
                self.__rear = -1
            else:
                self.__front = (self.__front + 1) % self.__max_size
            return data
    def display(self):
        if (self.is_empty()):
            print ("Queue is empty...")
        else:
            if (self.__front <= self.__rear):
                limit = self.__rear
            else:
                limit = self.__max_size - 1
            for location in range(self.__front, limit + 1):
                print (self.__elements[location], end = ", ")
            if (self.__front > self.__rear):
                for location in range(0, self.__rear + 1):
                    print (self.__elements[location], end = "; ")
        print ("")
    def __str__(self):
        if (self.is_empty()):
            msg = "Queue is empty..."
        else:
            if (self.__front <= self.__rear):
                limit = self.__rear
            else:
                limit = self.__max_size - 1
            msg = []
            for location in range(self.__front, limit + 1):
                msg.append((str)(self.__elements[location]))
            if (self.__front > self.__rear):
                for location in range(0, self.__rear + 1):
                    msg.append((str)(self.__elements[location]))
            msg = " ".join(msg)
            msg = "Queue content (Front to Rear): " + msg
        return msg
myqueue = Queue(6)
myqueue.enqueue(11); myqueue.enqueue(22); myqueue.enqueue(33)
myqueue.enqueue(44); myqueue.enqueue(55)
myqueue.display()
myqueue.enqueue(66); myqueue.enqueue(77)
myqueue.display()
print(myqueue.dequeue()); print(myqueue.dequeue())
print(myqueue.dequeue())
myqueue.display()
myqueue.enqueue(77); myqueue.enqueue(88)
myqueue.display()
print (myqueue)
```
